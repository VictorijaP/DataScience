### Завдання 2

Програма приймає на вхід два параметри: стать спортсмена(`sex = (f, m)`) та вік спортсмена(`age = [0,100]`).  Спортсмен-дівчина може брати участь у змаганні, якщо її вік більший 18, спортсмен-хлопець - якщо вік більший 16

Якщо введеніть не коректні дані, виведіть повідомлення про помилку

```python
sex = input('Введите пол спортсмена m или f: ')
age = float(input('Введите возраст спортсмена:'))
acceptable_sex = ['m', 'f']    
acceptable_age = range(1, 100)
if sex in acceptable_sex and age in acceptable_age:
    if sex == 'f' and age > 18:
        print('Участвует в соревнованиях')
    elif sex == 'm' and age > 16:
        print('Участвует в соревнованиях')
    else:
        print('Не участвует в соревнованиях')
else:
    print('Ошибка ввода данных')
 
```

Напишіть функції, що:

* медіану послідовності
* частоту кожної цифри в послідовності
* середнє квадратичне відхилення

Послідовність чисел ганеруємо рандомно

```python
import random
import math
from functools import reduce

#генеруэмо тисячу чисел в діапазоні від 0 до 10
list_with_random_numbers = [random.randint(0,100) for i in range(1000)]

max_val = reduce(max, list_with_random_numbers)
print('Максимальное значение:', max_val)

min_val = reduce(min, list_with_random_numbers)
print('Минимальное значение:', min_val)

middle_sum = sum(list_with_random_numbers) / len(list_with_random_numbers)
print('Среднее арифметическое:', middle_sum)

if len(list_with_random_numbers) % 2 == 0:
    ind1 = len(list_with_random_numbers) // 2 - 1
    ind2 = len(list_with_random_numbers) // 2 + 1
    mediana = (list_with_random_numbers[ind1] + list_with_random_numbers[ind2]) / 2
else:
    ind = len(list_with_random_numbers) / 2
    mediana = list_with_random_numbers[ind1]
    
print('Медиана:', mediana)

dict_with_numbers = {}
for el in list_with_random_numbers:
    count_el = list_with_random_numbers.count(el)
    dict_with_numbers[el] = count_el
print('Частота цифр в последовательности:', *dict_with_numbers.items(), sep='\n')

tmp = sum((i-middle_sum)**2 for i in list_with_random_numbers) / len(list_with_random_numbers)
st_dev = math.sqrt(tmp)
print('Среднеквадратичное отклонение:', st_dev)
```
